# Scraping <span style="color:red">bookshop.org</span>

```python
# import some important libraries for using in this scraping
from bs4 import BeautifulSoup as bs
import time
import undetected_chromedriver as uc
import numpy as np
```

```python
driver = uc.Chrome()
```

```python
# Extract all links of each page of the website below
pages = []
index = np.arange(1,101)
for i in index:
    pages.append("https://bookshop.org/categories/m/popular-books?page="+str(i))
print(pages)
```

```python
# Exrtract the links of each book in each page
book_links = []

def getBooks(web): # define function for getting book link
    #get the web page with our webdriver
    driver.get(web)
    #add a bit delay for the page to load properly before start scraping link
    time.sleep(1)
    #get the html for bs
    web = bs(driver.page_source, 'html.parser')
    #start looking for the 'a' tag name which store the href
    classs = web.find_all('h2', class_="font-serif-bold lg:pt-4 leading-tight")
    for link in classs:
        href = link.find('a').get('href')
        if href and '/p/books/' in href:
            book_links.append(href)

# Loop through all pages            
for web in pages:
    getBooks(web)
print(book_links)
    
```

```python
print("before duplicate : "+str(len(book_links)))
#To prevent any duplicate
books = []
[books.append(x) for x in book_links if x not in books]
print("after duplicate : "+str(len(books)))
```

```python
#add full http to the book links
book_last = []
for i in books:
    book_last.append("https://bookshop.org"+i)
print(book_last)
print(len(book_last))
```

```python
#The variable we will scrap include:
names = []
prices = []
#ratings = []
formats = []
pages = []
dimensions = []
pub_date = []
categories = []
#weights = []
languages = []
publishers = []
```

```python
#function for getting the data
def getData(web):
    driver.get(web)
    time.sleep(1)
    web = bs(driver.page_source,'html.parser')
    
    try:
        prices.append(web.find('div', itemprop="offers").select_one('b').text.replace('$',""))
    except:
        prices.append(None)
    try:
        publishers.append(web.find('div', itemprop="publisher").text)
    except:
        publishers.append(None)
    try:
        pub_date.append(web.find('div', itemprop="datePublished").text)
    except:
        pub_date.append(None)
    try:
        pages.append(web.find('div', itemprop="numberOfPages").text)
    except:
        pages.append(None)
    try:
        dimensions.append(web.find('div', itemprop="size").text)
    except:
        dimensions.append(None)    
    try:
        languages.append(web.find('div', itemprop="inLanguage").text)
    except:
        languages.append(None)
    try:
        formats.append(web.find('div', itemprop="bookFormat").text)
    except:
        formats.append(None)
    try:
        findAll_Category = web.find_all('a', itemprop="genre")
        category = ""
        for l in findAll_Category:
            category = category + l.text + "   "
        categories.append(category)
    except:
        categories.append(None)
    try:
        names.append(web.find('div', class_="mb-4").select_one('h1',class_="h1 leading-tight mb-2").text)
    except:
        names.append(None)
for i in book_last:
    getData(i)
```

```python
# Print the length of each features to make sure that they have the same length
print(len(names))
print(len(prices))
print(len(categories))
print(len(dimensions))
print(len(categories))
print(len(pages))
print(len(pub_date))
print(len(formats))
print(len(publishers))
```

```python
import pandas as pd
```

```python
# Create a dataframe of the scraped features
bookData = []
for i in range(0,len(prices)):
    bookData.append([names[i],prices[i],pages[i],formats[i],languages[i],pub_date[i],publishers[i],dimensions[i],categories[i]])
df = pd.DataFrame(data=bookData,columns=['Name','Prices','Page','Format','Language','Pub Date','Publisher','Dimensions','Categories'])
```
